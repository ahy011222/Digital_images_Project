```python
img1 = cv2.imread('/content/gdrive/MyDrive/digital/Images/lena256.jpg')
RGB_img = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)

gimg1 = cv2.cvtColor(RGB_img, cv2.COLOR_RGB2GRAY)
print(gimg1.shape)
cv2_imshow(gimg1)
#plt.imshow(gimg1, cmap='gray')
```

```python
gOutimg = np.zeros((gimg1.shape[0], gimg1.shape[1]))
```

```python
for h in range(0, gimg1.shape[0]):
  for w in range(0, gimg1.shape[1]):
    val = gimg1[h, w] % 64
    val = val * 4
    gOutimg[h, w] = val

#plt.imshow(gOutimg, cmap='gray')
cv2_imshow(gOutimg)
```

```python
equalVal = 100
widhth = 5

min = equalVal - widhth
max = equalVal + widhth
```

```python
for h in range(0, gimg1.shape[0]):
  for w in range(0, gimg1.shape[1]):
    if(gimg1[h,w] > min and gimg1[h,w] < max):
      gOutimg[h, w] = 0
    else:
      gOutimg[h, w] = gimg1[h,w]

plt.imshow(gOutimg, cmap='gray')
```

```python
for h in range(0, gimg1.shape[0]):
  for w in range(0, gimg1.shape[1]):
    val = 255*(gimg1[h,w]/128 -1)*(gimg1[h,w]/128 -1)
    gOutimg[h, w] = val

cv2_imshow(gOutimg)
```

```python
planeNumber = 1
planeValue = 1

for i in range(planeNumber -1):
  planeValue = planeValue * 2
```

```python
for h in range(0, gimg1.shape[0]):
  for w in range(0, gimg1.shape[1]):
    imVal = (int)(gimg1[h,w] / planeValue)
    if(imVal % 2 == 0):
      gOutimg[h, w] = 0
    else:
      gOutimg[h, w] = 255
cv2_imshow(gOutimg)
```
